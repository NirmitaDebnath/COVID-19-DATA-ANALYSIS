# COUNT OF NUMBER OF COUNTRIES UNDER WHO REGIONS

```python
ds=df['WHO Region'].value_counts()
ds.plot.pie(autopct='%.0f%%',subplots=True, figsize=(8,6),shadow=True, startangle=90,colors=sns.color_palette("YlOrBr"))
plt.title("WHO REGION AND THE % OF COUNTRIES UNDER THE REGION")
```

# COUNTRIES WITH MAXIMUM COVID CASES - 50

```python
plt.figure(figsize=(12,4))
ds=df.sort_values(by='Confirmed', ascending=False)
ds=ds.head(50)
ax=sns.barplot(y=ds['Confirmed'],x=ds['Country/Region'],data=ds,palette="cubehelix",linewidth=1,edgecolor="black")
plt.xticks(rotation=90)
ax.set(ylabel='CONFIRMED CASES', xlabel='COUNTRIES')
plt.title("50 COUNTRIES WITH MAXIMUM COVID CASES")
ax.set_facecolor("#EDB120")
```

# COUNTRIES WITH MINIMUM COVID CASES - 50

```python
plt.figure(figsize=(12,4))
ds=df.sort_values(by='Confirmed', ascending=True)
ds=ds.head(50)
ax=sns.barplot(y=ds['Confirmed'],x=ds['Country/Region'],data=ds,palette="Blues",linewidth=1.5,edgecolor="black")
plt.xticks(rotation=90)
ax.set(ylabel='CONFIRMED CASES', xlabel='COUNTRIES')
plt.title("50 COUNTRIES WITH MINIMUM COVID CASES")
ax.set_facecolor('#d0bbff')
```
